```python
import pandas as pd
from scipy.fftpack import fft
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as s
import cmath
import os
from scipy import stats 

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
print(color.BOLD + 'Activity in block E1: ' + color.END)
#directory = 'D:\FILES\linux-80211n-csitool-supplementary-master\linux-80211n-csitool-supplementary-master\matlab\CSV_files'
csv1 = 'no_act/csi_noactivity_1.csv'
csv2 = 'act/csi_walk_e1_1.csv'

df1 = pd.read_csv(csv1,usecols=range(1,91))
df1 =df1.applymap(lambda s: np.complex(s.replace('i', 'j')))
df2 = pd.read_csv(csv2,usecols=range(1,91))
df2 =df2.applymap(lambda s: np.complex(s.replace('i', 'j')))




# creating a list of dataframe columns 
columns = list(df1)

IQR1_a = []
IQR2_a = []
IQR1_b = []
IQR2_b = []
IQR1_c = []
IQR2_c = []
for i in range(1, 31):
    Z = []
    for k in ['a_']:
        Z.append(k+str(i))
        a_no = df1[Z[0]].values
        a = df2[Z[0]].values
        IQR1_a.append(stats.iqr(abs(a_no), interpolation = 'midpoint'))
        IQR2_a.append(stats.iqr(abs(a), interpolation = 'midpoint'))
    Z = []
    for k in ['b_']:
        Z.append(k + str(i)) 
        b_no = df1[Z[0]].values
        b = df2[Z[0]].values
        IQR1_b.append(stats.iqr(abs(b_no), interpolation = 'midpoint'))
        IQR2_b.append(stats.iqr(abs(b), interpolation = 'midpoint'))
    Z = []
    for k in ['c_']:
        Z.append(k + str(i)) 
        c_no = df1[Z[0]].values
        c = df2[Z[0]].values
        IQR1_c.append(stats.iqr(abs(c_no), interpolation = 'midpoint'))
        IQR2_c.append(stats.iqr(abs(c), interpolation = 'midpoint'))

        
            
        







quartile = np.arange(1,31)
plt.plot(quartile,IQR1_a,label = 'No Activity_a')
plt.plot(quartile,IQR2_a,label = 'Activity_a')
plt.legend()
plt.show()
plt.plot(quartile,IQR1_b,label = 'No Activity_b')
plt.plot(quartile,IQR2_b,label = 'Activity_b')
plt.legend()
plt.show()
plt.plot(quartile,IQR1_c,label = 'No Activity_c')
plt.plot(quartile,IQR2_c,label = 'Activity_c')
plt.legend()
plt.show()
```
