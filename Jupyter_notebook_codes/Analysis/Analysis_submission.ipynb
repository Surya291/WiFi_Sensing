```python
import pandas as pd
from scipy.fftpack import fft
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as s
import cmath
import os
from scipy import stats
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
print(color.BOLD + 'Analysis of CSI Amplitude variance' + color.END)
directory = 'D:\FILES\linux-80211n-csitool-supplementary-master\linux-80211n-csitool-supplementary-master\matlab\CSV_files'

print(color.CYAN + 'Plotting the variation of variance of amplitude for csi_noactivity' +color.END)
csv1 = directory+'\csi_noactivity_1.csv'
df1 = pd.read_csv(csv1,usecols=range(1,91))
df1 =df1.applymap(lambda s: np.complex(s.replace('i', 'j')))

V1 = np.zeros((30,3))
V_no = np.zeros(30)

for i in range(1, 31):
    Z1 = []
    for k in ['a_','b_','c_']:
        Z1.append(k + str(i)) 
    #print(Z)
    a1= df1[Z1[0]].values
    b1= df1[Z1[1]].values
    c1= df1[Z1[2]].values
    
    m_a1 = np.mean(abs(a1))
    m_b1 = np.mean(abs(b1))
    m_c1 = np.mean(abs(c1))
    
    var = [np.var((abs(a1)-m_a1)),np.var((abs(b1)-m_b1)),np.var((abs(c1)-m_c1))]
    V1[i-1][:] = var
    
V_no= V1[:,2] 
n = np.arange(1,31)
plt.plot(n,V_no[:],label = 'No Activity')
plt.legend()
plt.show()
```

```python
print(color.CYAN + 'Comparing the Variances obtained in No activity case with Variances of Activity' + color.END)
import os
directory = 'D:\FILES\linux-80211n-csitool-supplementary-master\linux-80211n-csitool-supplementary-master\matlab\CSV_files'
z = 'D:\FILES\linux-80211n-csitool-supplementary-master\linux-80211n-csitool-supplementary-master\matlab\CSV_files\\'

for filename in os.listdir(directory):
    if (filename.endswith(".csv") and filename.startswith("csi_walk")):
        a = z + filename
        print('Under Progress',filename)
        csv = a
        df = pd.read_csv(csv,usecols=range(1,91))
        df =df.applymap(lambda s: np.complex(s.replace('i', 'j')))
        V = np.zeros((30,1))
        
        
        for i in range(1, 31):
            Z = []
            for k in ['b_']:
                Z.append(k + str(i)) 

                b= df[Z[0]].values
                #print(np.shape(b))
                m_b = np.mean(abs(b))
                V[i-1] = np.var((abs(b)-m_b)) 
            #print(V[i-1])
        
        n = np.arange(1,31)
        
        plt.plot(n,V,label = 'Activity')
        plt.plot(n,V_no[:],label = 'No Activity')
        
        plt.show()
        
        
print('Mission Accomplished !!')
```

```python
print(color.BOLD + 'Analysis of CSI Amplitude Interquartile range' + color.END)
directory = 'D:\FILES\linux-80211n-csitool-supplementary-master\linux-80211n-csitool-supplementary-master\matlab\CSV_files'
print(color.CYAN + 'Plotting the variation of Interquartile range of amplitude for csi_noactivity' +color.END)

csv2 = directory+'\csi_noactivity_1.csv'
df2 = pd.read_csv(csv2,usecols=range(1,91))
df2 =df2.applymap(lambda s: np.complex(s.replace('i', 'j')))


No_activity_b = []
count = 0
No_act_quartile = []

for i in range(1, 31):
    Z2 = []
    for k in ['a_','b_','c_']:
        Z2.append(k + str(i)) 
    #print(Z)
    a2= df2[Z2[0]].values
    b2= df2[Z2[1]].values
    c2= df2[Z2[2]].values
    

    No_activity_b.append(stats.iqr(abs(b2), interpolation = 'midpoint'))

    count = count+1
    No_act_quartile.append(count)
    

plt.plot(No_act_quartile,No_activity_b,label = 'No Activity')
plt.legend()
plt.show()
```

```python
print(color.CYAN + 'Comparing the Interquartile range obtained in No activity case with Interquartile ranges  of Activity' + color.END)
directory = 'D:\FILES\linux-80211n-csitool-supplementary-master\linux-80211n-csitool-supplementary-master\matlab\CSV_files'
z = 'D:\FILES\linux-80211n-csitool-supplementary-master\linux-80211n-csitool-supplementary-master\matlab\CSV_files\\'

for filename in os.listdir(directory):
    if (filename.endswith(".csv") and filename.startswith("csi_walk")):
        a = z + filename
        print('Under Progress',filename)
        csv = a
        df = pd.read_csv(csv,usecols=range(1,91))
        df =df.applymap(lambda s: np.complex(s.replace('i', 'j')))
        V = []
        
        
        for i in range(1, 31):
            Z = []
            for k in ['b_']:
                Z.append(k + str(i)) 

                b= df[Z[0]].values
                #print(np.shape(b))
                V.append(stats.iqr(abs(b), interpolation = 'midpoint'))
            
        
        n = np.arange(1,31)
        
        plt.plot(n,V,label = 'Activity')
        plt.plot(n,No_activity_b,label = 'No activity')
        plt.legend()
        plt.show()
        
print('Mission Accomplished !!')
```
